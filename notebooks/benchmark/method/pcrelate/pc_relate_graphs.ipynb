### Speedup

```python
fixed_variants_2k_block_1_core = pd.read_csv(Experiment("docker_fixed_variants_2k_cpu_8_ram_31_blocksize_1_core").experiment_file)


def speedup(i: pd.DataFrame, baseline: pd.DataFrame = fixed_variants_2k_block_1_core) -> pd.DataFrame:
    s = baseline.samples.to_frame()
    s["pc_relate_speedup"] = baseline.pc_relate/i.pc_relate
    return s

p = sns.scatterplot(x="samples", y="pc_relate_speedup", data=speedup(fixed_variants_2k_block_1_core),s=dot_size)
p = sns.scatterplot(x="samples", y="pc_relate_speedup", data=speedup(fixed_variants_2k_block_2_cores_d), s=dot_size, ax=p)
p = sns.scatterplot(x="samples", y="pc_relate_speedup", data=speedup(fixed_variants_2k_block_3_cores_d), s=dot_size, ax=p)
p = sns.scatterplot(x="samples", y="pc_relate_speedup", data=speedup(fixed_variants_2k_block_4_cores_d), s=dot_size, ax=p)
p = sns.scatterplot(x="samples", y="pc_relate_speedup", data=speedup(fixed_variants_2k_block_cpu), s=dot_size, ax=p)
p.set(ylabel=common_y_label, title="variants=2000, MAF=0")
p.legend(["block_size=samples", "block_size=2cores", "block_size=3cores", "block_size=4cores", "block_size=8cores", "block_size=512"])
```

```python
import numpy as np

def fit_quadratic(data: pd.DataFrame) -> pd.DataFrame:
    params = np.polyfit(data.samples, data.pc_relate, 2)
    return np.poly1d(params)

one_core = fit_quadratic(fixed_variants_2k_block_samples)
eight_cores = fit_quadratic(fixed_variants_2k_block_cpu)
p = sns.scatterplot(x="samples", y="pc_relate", data=fixed_variants_2k_block_samples, s=dot_size)
p = sns.scatterplot(x="samples", y="pc_relate", data=fixed_variants_2k_block_cpu, s=dot_size)
plt.plot(fixed_variants_2k_block_samples.samples, one_core(fixed_variants_2k_block_samples.samples))
plt.plot(fixed_variants_2k_block_cpu.samples, eight_cores(fixed_variants_2k_block_cpu.samples))
print(f"0.5M samples (AND 2000 variants) would take {one_core(.5e6)/3600/24:.2f} days on a single core")
print(f"0.5M samples (AND 2000 variants) would take {eight_cores(.5e6)/3600/24:.2f} days on 8 cores")
```

```python
from toolz import partial

def amdahls_law(s: int, p:float) -> float:
    return 1/((1-p)+p/s)

x = np.array([2 ** x for x in range(1,17)])
p = sns.lineplot(x=x, y=partial(amdahls_law, p=.8)(x))
p = sns.lineplot(x=x, y=partial(amdahls_law, p=.95)(x), ax=p)
p.set(xscale="log")
```

```python
from scipy.optimize import curve_fit

def get_speedup_at(data: pd.DataFrame, baseline: pd.DataFrame = fixed_variants_2k_block_1_core, sample: int = 4000):
    s = speedup(data, baseline)
    ind = s.loc[s.samples == sample].index.item()
    assert s.at[ind, "samples"] == sample    
    return s.at[ind, "pc_relate_speedup"]


s = [get_speedup_at(d) for d in [fixed_variants_2k_block_1_core,
                                 fixed_variants_2k_block_2_cores_d,
                                 fixed_variants_2k_block_3_cores_d,
                                 fixed_variants_2k_block_4_cores_d,
                                 fixed_variants_2k_block_cpu]]

popt, pcov = curve_fit(amdahls_law, [1,2,3,4,8], s, 0.8)
print(f"popt: {popt[0]}")
x = np.linspace(1, 100, 100)
p = sns.lineplot(x=x, y=partial(amdahls_law, p=popt[0])(x))
p.set(ylabel="speedup", xlabel="number of cores")
```

```python
pc_relate_speedup = partial(amdahls_law, p=popt[0])
```

```python
print(f"From Amdahl's law: {one_core(10000)/pc_relate_speedup(8):.2f} seconds")
print(f"From on 8 core times fit: {eight_cores(10000):.2f} seconds")
print(f"Actual time: {fixed_variants_2k_block_cpu.loc[fixed_variants_2k_block_cpu.samples == 10000, 'pc_relate'].iloc[0]} seconds")
```

```python
def pc_relate_time(X, a, b, c, d, e):
    s,v = X
    return (a * v + b) * (c * s**2 + d * s + e)

def fit_pc_relate_comp(e: pd.DataFrame):
    popt, _ = curve_fit(pc_relate_time, (e.samples, e.variants), e.pc_relate)
    return lambda s, v: pc_relate_time((s, v), *popt)


cpu_4 = get_cpu_experiment(4)
cpu_ex_1_core = fit_pc_relate_comp(cpu_1)
cpu_ex_4_core = fit_pc_relate_comp(cpu_4)
p = sns.scatterplot(x="samples", y="pc_relate", data=cpu_1, s=dot_size)
p = sns.scatterplot(x="samples", y="pc_relate", data=cpu_4, s=dot_size)
plt.plot(cpu_1.samples, cpu_ex_1_core(cpu_1.samples, cpu_1.variants))
plt.plot(cpu_4.samples, cpu_ex_4_core(cpu_4.samples, cpu_4.variants))
print(f"0.5M samples and 2.5M variants would take {cpu_ex_1_core(.5e6, 2.5e6)/3600/24:.2f} days on a single core")
print(f"0.5M samples and 2.5M variants would take {cpu_ex_4_core(.5e6, 2.5e6)/3600/24:.2f} days on 4 cores")
```
